```python
import json
from collections import defaultdict

fFile1 = "/home/nebulamind/Documents/AI Lab/Scraping/tags/first_file_given_final.json"
fFile2 = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quran_classified_cleaned.json"
fFile3 = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quran_topics_with_verses_cleaned2.json"
fFile4 = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quranindex_info_cleaned.json"

out1 = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quran_tags_final.json"

# Load all files
with open(fFile1) as f1, open(fFile2) as f2, open(fFile3) as f3, open(fFile4) as f4:
    data1 = json.load(f1)
    data2 = json.load(f2)
    data3 = json.load(f3)
    data4 = json.load(f4)

# Prepare a dictionary to collect all data grouped by verse ID
merged = defaultdict(lambda: {
    "id": "",
    "name": None,
    "data_category": None,
    "main_Topic1": None,
    "subtopic1": [],
    "mainTopic2": None,
    "subtopic2": None,
    "mainTopic3": None,
    "subtopic3": []
})

# File 1: Add name, data_category, main_Topic1, subtopic1[0]
for item in data1:
    verse_id = item.get("ayah_id")
    if verse_id:
        merged[verse_id]["id"] = verse_id
        merged[verse_id]["name"] = item.get("name")
        merged[verse_id]["data_category"] = item.get("data_category")
        merged[verse_id]["main_Topic1"] = item.get("Topic Title")
        if item.get("Sub-Topic"):
            merged[verse_id]["subtopic1"].append(item["Sub-Topic"])

# File 4: Add topic from file4 as subtopic1[1]
for item in data4:
    verse_id = item.get("verse")
    if verse_id:
        merged[verse_id]["id"] = verse_id
        merged[verse_id]["subtopic1"].append(item["topic"])

# File 2: Add chapter/subtag (mainTopic2/subtopic2)
for item in data2:
    verse_id = item.get("surah_info")
    if verse_id:
        merged[verse_id]["id"] = verse_id
        merged[verse_id]["mainTopic2"] = item.get("chapter")
        merged[verse_id]["subtopic2"] = item.get("subtag")

# File 3: Add subtopics
for item in data3:
    verse_id = item.get("verses")
    if verse_id:
        subtopics = item.get("subtopics", [])
        if subtopics:
            merged[verse_id]["id"] = verse_id
            merged[verse_id]["mainTopic3"] = subtopics[0]
            if len(subtopics) > 2:
                merged[verse_id]["subtopic3"] = subtopics[1:-1]

# Convert to list and save
output = list(merged.values())

# Save output to final.json
with open(out1, "w", encoding="utf-8") as out:
    json.dump(output, out, ensure_ascii=False, indent=2)

print("✅ Merged data saved to 'merged_output.json'")
```

```python
import json


input = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quran_tags_final_with_translations.json"
output = "/home/nebulamind/Documents/AI Lab/Scraping/tags/quran_tags_final_with_translations_structured.json"

# Load the file (assuming single object or list of objects)
import json

def deduplicate_case_insensitive(items):
    seen = set()
    result = []
    for item in items:
        key = item.lower()
        if key not in seen:
            seen.add(key)
            result.append(item)
    return result

# Load JSON data
with open(input, "r", encoding="utf-8") as f:
    data = json.load(f)

# Ensure data is a list
if isinstance(data, dict):
    data = [data]

processed = []
for item in data:
    # Combine main topics safely
    main_topic_parts = [
        item.get("name", ""),
        item.get("mainTopic1", ""),
        item.get("main_Topic2", ""),
        item.get("mainTopic3", "")
    ]
    main_topic = deduplicate_case_insensitive(
        [part for part in main_topic_parts if isinstance(part, str) and part.strip()]
    )

    # Combine subtopics safely
    subtopic_parts = []

    subtopic_parts.extend(item.get("data_category") or [])
    subtopic1 = item.get("subtopic1")
    if subtopic1: subtopic_parts.append(subtopic1)
    subtopic_parts.extend(item.get("subtopic2") or [])
    subtopic_parts.extend(item.get("subtopic3") or [])

    # Filter non-empty string entries
    subtopic = deduplicate_case_insensitive(
        [s for s in subtopic_parts if isinstance(s, str) and s.strip()]
    )

    processed.append({
        "id": item["id"],
        "translation": item.get("translation", ""),
        "main_topic": main_topic,
        "subtopic": subtopic
    })

# Write to output JSON
with open(output, "w", encoding="utf-8") as f_out:
    json.dump(processed, f_out, indent=2, ensure_ascii=False)
```

```python
import pandas as pd
import json
from collections import Counter

# --- File paths ---
translation_file = "/media/nebulamind/AI Lab/AI Lab/AskDeen/dataset/Translation File(2).xlsx"
json_file = "/media/nebulamind/AI Lab/AI Lab/Scraping/tags/quran_tags_final_with_translations_structured1.json"

# --- Load data ---
with open(json_file, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

df = pd.read_excel(translation_file)
excel_ids = set(df['ID'].astype(str))

# --- Collect JSON IDs ---
json_ids = [item.get('id') for item in json_data]

# --- Check for duplicates ---
id_counts = Counter(json_ids)
duplicates = [id for id, count in id_counts.items() if count > 1]

if duplicates:
    print("🔁 Duplicate IDs found in JSON:\n")
    for dup_id in duplicates:
        print(f"--- ID: {dup_id} (occurs {id_counts[dup_id]} times) ---")
        for entry in json_data:
            if entry.get('id') == dup_id:
                print(json.dumps(entry, indent=2, ensure_ascii=False))
        print("\n" + "-"*60 + "\n")
else:
    print("✅ No duplicate IDs found in JSON.")

# --- Check for IDs in JSON but not in Excel ---
missing_in_excel = [entry for entry in json_data if entry.get('id') not in excel_ids]

if missing_in_excel:
    print("\n❌ IDs in JSON but NOT found in Excel:\n")
    for entry in missing_in_excel:
        print(f"--- ID: {entry.get('id')} ---")
        print(json.dumps(entry, indent=2, ensure_ascii=False))
        print("\n" + "-"*60 + "\n")
    print(f"Total missing in Excel: {len(missing_in_excel)}")
else:
    print("\n✅ All JSON IDs are present in Excel.")
```

```python
import json
import pandas as pd

# --- File Paths ---
json_file = "/media/nebulamind/AI Lab/AI Lab/Scraping/required_output/quran_tags_final_with_translations_structured1.json"  # Replace with your JSON file path
excel_file = "/media/nebulamind/AI Lab/AI Lab/AskDeen/dataset/Translation File(2).xlsx"  # Replace with your Excel file path
output_file = "/media/nebulamind/AI Lab/AI Lab/Scraping/required_output/quran_tags_final_with_translations_structured1.json"

# --- Load JSON ---
with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# --- Load Excel ---
df = pd.read_excel(excel_file)
df['ID'] = df['ID'].astype(str)  # Ensure ID column is string
id_to_arabic = dict(zip(df['ID'], df['Arabic']))  # Create a mapping: ID → Arabic

# --- Function to extract (surah, ayah) as tuple for sorting ---
def id_sort_key(item):
    try:
        surah, ayah = map(int, item['id'].split('|'))
        return (surah, ayah)
    except:
        return (999, 999)  # Put malformed IDs at the end

# --- Add Arabic and sort ---
for item in data:
    item_id = item.get('id')
    item['arabic'] = id_to_arabic.get(item_id, '')  # Add Arabic or empty if not found

sorted_data = sorted(data, key=id_sort_key)

# --- Save to file ---
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(sorted_data, f, indent=2, ensure_ascii=False)

print(f"✅ JSON sorted and Arabic added. Output saved to: {output_file}")
```
